```python
import glob
output = np.sort(glob.glob('./Outputs/*.csv'))

_dict = pd.read_csv(output[0], header=None, index_col=0, squeeze=True).to_dict()
# for key in _dict:
#     print(key, float(_dict[key].replace("[","").replace("]","")))

def unmangle(x):
    
    try:
        x = x.lstrip()
        x = x.rstrip()
        x = x.lstrip("[")
        x = x.rstrip("]")
        x = x.lstrip()
        x = x.rstrip()
        x = x.split(None)

        xf = np.array([float(xi) for xi in x])
    except:
        print("Can not unmangle", x)
    
    return xf


unmangle(_dict['u_quadlimbdark__'])
```

```python
kic = []
per = []
ror = []
b   = []
dur = []
rhocirc = []

for out in output:
    _dict    = pd.read_csv(out, header=None, index_col=0, squeeze=True).to_dict()
    _kic     = int(out.split("KIC")[1][:-4])
    
    _per     = unmangle(_dict["period"])
    _ror     = unmangle(_dict["ror"])
    _b       = unmangle(_dict["b"])
    _dur     = unmangle(_dict["dur"])
    _rhocirc = unmangle(_dict["rho_circ"])
    
    for i in range(len(_per)):
        kic.append(_kic)
        per.append(_per[i])
        ror.append(_ror[i])
        b.append(_b[i])
        dur.append(_dur[i])
        rhocirc.append(_rhocirc[i])
        
        
        
s21 = pd.DataFrame({"kic": kic,
                    "per": per,
                    "ror": ror,
                    "b": b,
                    "dur": dur,
                    "rhocirc": rhocirc})

print(np.shape(s21))
```

```python
s21 = s21.merge(new_df, how="left", left_on="kic", right_on="dr25_kepid")
```

```python
import astropy.constants as c

s21["rp"] = s21["ror"]*s21["b20_iso_rad"]*c.R_sun/c.R_earth
```

```python
plt.scatter(s21["per"], s21["dur"])
plt.semilogy()
plt.show()
#plt.hist(s21["rp"], bins=50)

plt.scatter(s21["per"], s21["rp"], c=s21["dur"])
plt.semilogx()
plt.gcf().set_size_inches(8,8)
plt.show()

plt.scatter(s21["per"], s21["rp"], c=s21["rhocirc"])
plt.semilogx()
plt.gcf().set_size_inches(8,8)
plt.show()
```
